```python
# Feature Engineering
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df['gdp_total']=df['gdp_per_capita'] * df['population']
```

```python
df['gdp_total'].head()
```

```python
df['fdi_per_capita']=(df['fdi_inflows'] * 1_000_000_000) /df['population']
df['fdi_per_capita']
```

```python
df['fdi_per_capita'].describe()
```

```python
print(df['fdi_inflows'].describe())
print(df['population'].describe())
```

```python
df['health_index']=(df['life_expectancy']/100) + (1/df['infant_mortality'])
```

```python
df['education_index']=(df['school_enrolment_primary'] + df['school_enrolment_secondary'])/2
```

```python
df['environment_index']=(df['renewable_energy'] + df['forest_area'] + df['clean_water_access'])/3
```

```python
df['gender_index']=(df['female_labour'] + df['women_parliament'] + df['gender_parity']) / 3
```

```python
df['development_index']=(df['electricity_access'] + df['clean_water_access']+df['internet_users']+df['education_index'])/4
```

```python
df['development_category']=pd.cut(df['development_index'],
       bins=[0,30,50,70,90,100,110],
       labels=['Very Poor','Underdeveloped','Developing','Moderate','Developed','Highly Developed']
    )
```

```python
df.head()
```

```python
df['life_expectancy_category']=pd.cut(df['life_expectancy'],
       bins=[0,50,60,70,80,90],
       labels=['Very Low','Low','Moderate','High','Very High']
      )
```

```python
df['environment_category'] = pd.cut(
    df['environment_index'],
    bins=[0, 25, 40, 55, 70, 100],
    labels=['Critical', 'Poor', 'Moderate', 'Good', 'Excellent']
)
```

```python
df['gdp_category'] = pd.cut(
    df['gdp_per_capita'],
    bins=[0, 1000, 5000, 15000, 40000, 200000],
    labels=['Low Income', 'Lower Middle', 'Upper Middle', 'High Income', 'Very High Income']
)
```

```python
df['gender_category'] = pd.cut(
    df['gender_index'],
    bins=[0, 20, 40, 60, 80, 100],
    labels=['Very Poor', 'Poor', 'Moderate', 'Good', 'Excellent']
)
```
